```r
library(MAESTRO)
library(chromVAR)
library(Seurat)
library(Matrix)
library(BSgenome.Hsapiens.UCSC.hg38)
library(BiocParallel)
library(motifmatchr)
library(SummarizedExperiment)
library(rhdf5)
library(JASPAR2016)
register(MulticoreParam(8))
```

```r
chromVar <- function(peakfile, bamfile, celltype){
    peaks <- readNarrowpeaks(peakfile,non_overlapping = FALSE)
    fragment_counts <- getCounts(bamfile, 
                                 peaks, 
                                 paired =  TRUE, 
                                 by_rg = TRUE, 
                                 format = "bam", 
                                 colData = DataFrame(celltype = celltype))

    motifs <- getJasparMotifs()
    motif_ix <- matchMotifs(motifs, fragment_counts, 
                            genome = BSgenome.Hsapiens.UCSC.hg38)
    return(motif_ix)
}
```
